```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

# 1. Setup Data Parameters
states = ['Lagos', 'Ogun', 'Oyo', 'Osun', 'Ekiti', 'Ondo', 'Rivers', 'Bayelsa', 'Edo', 'Enugu', 'Imo', 'Anambra', 'Kano', 'Katsina', 'FCT', 'Benue']
models = {
    'Toyota': ['Corolla', 'Camry', 'Prado', 'Hilux', 'Venza', 'Sienna'],
    'Honda': ['Accord', 'Civic', 'CR-V', 'Pilot'],
    'Benz': ['C300', 'E350', 'C-Class', 'GLK 350', 'S550', 'GLE 450', 'GLC 300'],
    'Lexus': ['ES 350', 'RX 350', 'GX 460', 'IS 250'],
    'Hyundai': ['Elantra', 'Santa Fe', 'Tucson'],
    'Ford': ['Explorer', 'Edge', 'F-150']
}

# 2. Create the Raw Data (8,000 rows)
data = []
start_date = datetime(2021, 1, 1)

for i in range(8000):
    state = random.choice(states)
    # Assign Rep based on Centralized Lagos Hub logic
    if state in ['Lagos', 'Ogun', 'Oyo', 'Osun', 'Ekiti', 'Ondo']: rep = 'Olawale'
    elif state in ['Rivers', 'Bayelsa', 'Edo', 'Enugu', 'Imo', 'Anambra']: rep = 'Chukwudi'
    elif state in ['Kano', 'Katsina', 'FCT']: rep = 'Aminu'
    else: rep = 'Adeola'
    
    brand = random.choice(list(models.keys()))
    model = random.choice(models[brand])
    
    # Introduce "Messy" Data: Randomly misalign brand for cleaning practice
    if random.random() < 0.15: 
        display_brand = "Unassigned" 
    else: 
        display_brand = brand
        
    price = random.randint(12000000, 85000000)
    qty = random.randint(1, 3)
    date = start_date + timedelta(days=random.randint(0, 1850))
    
    data.append([date, state, rep, display_brand, model, price, qty, round(random.uniform(0, 20), 1)])

# 3. Create DataFrame
df_raw = pd.DataFrame(data, columns=['Sale_Date', 'State', 'Sales_Rep', 'Car_Brand', 'Car_Model', 'Unit_Price_Naira', 'Quantity', 'Discount_Percent'])

# 4. Save to CSV
df_raw.to_csv('Nigeria_Car_Sales_V2_RAW.csv', index=False)
print("File 'Nigeria_Car_Sales_V2_RAW.csv' has been generated in your folder!")
```
